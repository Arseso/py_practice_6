```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
trees = pd.read_csv('./data/trees.csv')
trees
```

```python
fig, ax = plt.subplots(len(trees.columns.values[1:]), 2, figsize=(10,7))

for i, col in enumerate(trees.columns.values[1:]):
    sns.scatterplot(data=trees, x="ID", y=col, ax=ax[i, 0])
    ax[i, 0].set_title(f"Scatter plot for {col}")

    sns.histplot(data=trees, x=col, ax=ax[i, 1])
    ax[i, 1].set_title(f"Histogram for {col}")

plt.tight_layout()
plt.show()
```
